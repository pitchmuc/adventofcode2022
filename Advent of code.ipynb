# Day 2

```python
with open('input_day2.txt','r') as f:
    data = f.readlines()
data = [line.strip() for line in data]
```

```python
matrix = [tuple(row.split(' ')) for row in data]
```

```python
matrix
```

* A : Rock
* B : Paper
* C : Scissor

Power
* A > C
* C > B
* B > A

```python
translator = {
    "X" : "A",
    "Y" : "B",
    "Z" : "C"
}
```

```python
def scoringStrategy(opponent, you):
    translatedYou = translator[you]
    score = 0
    ## hand score
    if translatedYou == "A":
        score += 1
    elif translatedYou == "B":
        score +=2
    elif translatedYou == "C":
        score += 3
    ## game score
    if translatedYou == opponent:
        score += 3
    elif translatedYou == "A" and opponent == "C":
        score += 6
    elif translatedYou == "C" and opponent == "B":
        score += 6
    elif translatedYou == "B" and opponent == "A":
        score += 6
    else:
        score += 0
    return score
```

```python
myResult = []
for row in matrix:
    myResult.append(scoringStrategy(row[0],row[1]))
```

```python
len(myResult)
```

```python
sum(myResult)
```

```python
def findMyHand(opponent, outcome):
    if outcome == 'X':
        if opponent == 'A':
            return 'C'
        if opponent == 'B':
            return 'A'
        if opponent == 'C':
            return 'B'
    if outcome == "Z":
        if opponent == 'A':
            return 'B'
        if opponent == 'B':
            return 'C'
        if opponent == 'C':
            return 'A'
    if outcome == 'Y':
        return opponent
        
```

```python
def scoringStrategy2(opponent, outcome):
    translatedYou = findMyHand(opponent,outcome)
    score = 0
    ## hand score
    if translatedYou == "A":
        score += 1
    elif translatedYou == "B":
        score +=2
    elif translatedYou == "C":
        score += 3
    ## game score
    if translatedYou == opponent:
        score += 3
    elif translatedYou == "A" and opponent == "C":
        score += 6
    elif translatedYou == "C" and opponent == "B":
        score += 6
    elif translatedYou == "B" and opponent == "A":
        score += 6
    else:
        score += 0
    return score
```

```python
myResult = []
for row in matrix:
    myResult.append(scoringStrategy2(row[0],row[1]))
```

```python
len(myResult)
```

```python
sum(myResult)
```
